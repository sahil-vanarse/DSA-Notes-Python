```python
def dfs(graph, startNode, visited=set()):
    """
    Perform Depth-First Search (DFS) traversal on a graph represented using an adjacency list.

    Args:
    - graph (dict): The adjacency list representation of the graph.
    - startNode (int): The starting node for DFS traversal.
    - visited (set): A set to keep track of visited nodes. Default is an empty set.

    Returns:
    - None
    """
    print(startNode)
    visited.add(startNode)

    for child in graph[startNode]:
        if child not in visited:
            dfs(graph, child, visited)

# Adjacency list representation of the graph
ipt = [[1,2], [1,5], [2,3], [2,4], [2,5], [3,4], [3,6], [4,5], [4,6], [5,6]]
graph = {
    1: [2, 5],
    2: [1, 3, 4, 5],
    3: [2, 4, 6],
    4: [2, 3, 5, 6],
    5: [1, 2, 4, 6],
    6: [3, 4, 5]
}

# DFS traversal starting from node 1
dfs(graph, 1)
```

```python
# Define a function to perform depth-first search (DFS) traversal on a graph
def dfs(graph, startNode, visited):
    # Print the current node being visited
    print(startNode)
    # Mark the current node as visited
    visited.add(startNode)
    # Initialize the number of components to 0
    number_of_components = 0
    # Traverse all adjacent nodes of the current node
    for child in graph[startNode]:
        # If the adjacent node has not been visited
        if child not in visited:
            # Recursively call dfs on the adjacent node and increment the number of components
            number_of_components += dfs(graph, child, visited)
    # After traversal, return the total number of components in the current connected component
    return number_of_components + 1
    

# Define edges between nodes
edges = [['A', 'B'], ['A', 'C'], ['A', 'D'], ['D', 'E'], ['B', 'E'], ['F', 'H'], ['F', 'G'], ['I', 'J']]
# Define nodes in the graph
nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K' ]

# Create an empty graph
graph = dict()

# Initialize the graph with empty lists for each node
for key in nodes:
    graph[key] = list()

# Populate the graph with edges
for v1, v2 in edges:
    graph[v1].append(v2)
    graph[v2].append(v1)

# Initialize a set to keep track of visited nodes
visited = set()

# Initialize an empty list to store the number of components in each connected component
answer = list()

# Iterate over all nodes in the graph
for node in nodes:
    # If the current node has not been visited yet
    if node not in visited:
        # Perform DFS from the current node and store the number of components in the connected component
        temp = dfs(graph, node, visited)
        # Append the number of components to the answer list
        answer.append(temp)

# Print the number of components in each connected component
print(answer)
```

```python
# Define a function for single source shortest path (SSSP) using DFS
def sssp(graph, node, d, distance, parent):
    # Update the distance of the current node from the source node
    distance[node] = d
    # Iterate over all adjacent nodes of the current node
    for child in graph[node]:
        # If the adjacent node is not the parent (to avoid going back)
        if child != parent:
            # Recursively call sssp for the adjacent node and update its distance
            sssp(graph, child, distance[node] + 1, distance, node)

# Define edges between nodes
edges = [['A', 'B'], ['A', 'C'], ['C', 'E'], ['C', 'F'], ['B', 'D'], ['D', 'G'], ['G', 'I'], ['G', 'H']]
# Define nodes in the graph
nodes = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I']

# Create an empty graph
graph = dict()

# Create a dictionary to store distances from the source node
distance = dict()
# Initialize distances to None for all nodes
for key in nodes:
    graph[key] = list()
    distance[key] = None

# Populate the graph with edges
for v1, v2 in edges:
    graph[v1].append(v2)
    graph[v2].append(v1)

# Define the source node
start = 'A'
# Compute the shortest paths from the source node using the sssp function
sssp(graph, start, 0, distance, -1)

# Print the distances from the source node to all other nodes
for key, value in distance.items():
    print(key, ' --> ', value)
```

```python
# Define a function to check if a graph is bipartite
def bipartite(graph, node, visited, color, c):
    # Mark the current node as visited
    visited[node] = 1
    # Assign color to the current node
    color[node] = c

    # Iterate over all adjacent nodes of the current node
    for child in graph[node]:
        # If the adjacent node has not been visited
        if visited[child] != 1:
            # Recursively call bipartite function for the adjacent node with a different color
            temp = bipartite(graph, child, visited, color, 1 - color[node])
            # If the recursion returns False, the graph is not bipartite, so return False
            if temp == False:
                return False
        else:
            # If the adjacent node has been visited and has the same color as the current node, the graph is not bipartite
            if color[node] == color[child]:
                return False
    # If all conditions pass, return True indicating the graph is bipartite
    return True

# Define edges of the graph
ipt = [[1,2], [2,4], [2,5], [4,3], [3,1], [4,5]]
# Define the number of nodes in the graph
n = 5

# Create an empty graph, visited dictionary, and color dictionary
graph = dict()
visited = dict()
color = dict()

# Initialize the graph, visited, and color dictionaries
for i in range(1, n+1):
    graph[i] = list()
    visited[i] = 0
    color[i] = None

# Populate the graph with edges
for v1, v2 in ipt:
    graph[v1].append(v2)
    graph[v2].append(v1)

# Call the bipartite function to check if the graph is bipartite, starting from node 1 with color 0
print(bipartite(graph, 1, visited, color, 0))

# Print the color of each node
print(color)
```

```python
# Define a function to detect cycles in a graph using depth-first search (DFS)
def cycleDetection(graph, node, visited, parent):
    # Mark the current node as visited
    visited[node] = True
    # Iterate over all adjacent nodes of the current node
    for child in graph[node]:
        # If the adjacent node has not been visited
        if visited[child] == False:
            # Recursively call cycleDetection function for the adjacent node
            # If cycle is detected in the subtree rooted at the adjacent node, return True
            if cycleDetection(graph, child, visited, node):
                return True
        else:
            # If the adjacent node has been visited and is not the parent, a cycle is detected
            if child != parent:
                return True
    # If no cycle is detected in the subtree rooted at the current node, return False
    return False

# Define edges of the graph
ipt = [[1,2], [1,3], [2,3], [2,4], [4,5], [5,1]]
# Define the number of nodes in the graph
n = 5

# Create an empty graph and visited dictionary
graph = dict()
visited = dict()

# Initialize the graph and visited dictionaries
for i in range(1, n+1):
    graph[i] = list()
    visited[i] = False

# Populate the graph with edges
for v1, v2 in ipt:
    graph[v1].append(v2)
    graph[v2].append(v1)

# Call the cycleDetection function to check if the graph contains a cycle, starting from node 1 with no parent
print(cycleDetection(graph, 1, visited, None))

# Print the visited dictionary and the graph
print(visited)
print(graph)
```

```python
# Define a function to perform depth-first search (DFS) and calculate in and out times of each node in a tree
def dfs(graph, node, visited, inTime, outTime):
    global timer
    # Set the in-time of the current node
    inTime[node] = timer
    timer += 1
    visited[node] = True

    # Iterate over all adjacent nodes of the current node
    for child in graph[node]:
        # If the adjacent node has not been visited
        if visited[child] == False:
            # Recursively call dfs function for the adjacent node
            dfs(graph, child, visited, inTime, outTime)
    # Set the out-time of the current node
    outTime[node] = timer
    timer += 1

# Define edges of the tree
ipt = [[1,2], [1,3], [1,8], [2,4], [2,5], [2,6], [3,7]]
# Define the number of nodes in the tree
n = 8

# Create an empty graph and visited dictionary
graph = dict()
visited = dict()
# Initialize timer, inTime, and outTime dictionaries
timer = 1
inTime = dict()
outTime = dict()

# Initialize the graph, visited, inTime, and outTime dictionaries
for i in range(1, n+1):
    graph[i] = list()
    visited[i] = False
    inTime[i] = None
    outTime[i] = None

# Populate the graph with edges
for v1, v2 in ipt:
    graph[v1].append(v2)
    graph[v2].append(v1)

# Call the dfs function to calculate in and out times, starting from node 1
dfs(graph, 1, visited, inTime, outTime)

# Print the in and out times of each node
print(inTime)
print(outTime)
```

```python
# Define a depth-first search (DFS) function to calculate the height of a subtree rooted at a given node
def dfs(graph, node, newVisited):
    visited[node] = True  # Mark the current node as visited
    height = 0
    for child in graph[node]:
        if not visited[child]:  # If the child node has not been visited
            # Recursively call dfs function for the child node and update the height
            height = max(dfs(graph, child, newVisited), height)
    return height + 1  # Return the height of the subtree rooted at the current node

# Define a function to find the deepest node in a tree and its height
def dfs1(graph, node, visited):
    visited[node] = True  # Mark the current node as visited
    height = 0
    temp = node  # Initialize the deepest node as the current node
    for child in graph[node]:
        if not visited[child]:  # If the child node has not been visited
            # Recursively call dfs1 function for the child node and update the height
            tvalue, tnode = dfs1(graph, child, visited)
            if tvalue > height:  # If the height of the subtree rooted at the child node is greater than the current height
                temp = tnode  # Update the deepest node
                height = tvalue  # Update the height
    return [height + 1, temp]  # Return the height of the subtree rooted at the current node and the deepest node

# Define the edges of the tree
ipt = [[1,2], [2,4], [2,5], [2,6], [1,3], [3,7], [1,8], [6,10], [1,9]]
n = 10  # Total number of nodes in the tree

graph = dict()
visited = dict()

# Initialize the graph and visited dictionaries
for i in range(1, n+1):
    graph[i] = list()  # Initialize an empty list for each node to store its adjacent nodes
    visited[i] = False  # Initialize each node as not visited

# Populate the graph with edges
for v1, v2 in ipt:
    graph[v1].append(v2)  # Add v2 to the list of adjacent nodes of v1
    graph[v2].append(v1)  # Add v1 to the list of adjacent nodes of v2

# Find the deepest node in the tree and its height
temp = dfs1(graph, 1, visited)  # Find the deepest node starting from the root (node 1)
nextNode = temp[1]  # Get the deepest node

newVisited = dict()

# Initialize the newVisited dictionary
for i in range(1, n+1):
    visited[i] = False  # Initialize each node as not visited

# Perform DFS starting from the deepest node to calculate the height of the subtree rooted at that node
dfs(graph, nextNode, newVisited)
```

```python
# Calculating SubTree

def SubTreeNodes(graph, node, visited, distance):
    visited[node] = True
    height = 0
    for child in graph[node]:
        if not visited[child]:
            height += SubTreeNodes(graph, child, visited, distance)
    distance[node] = height + 1
    return distance[node]

ipt = [[1,2], [2,4], [2,5], [2,6], [1,3], [3,7], [1,8], [6,10], [1,9]]
n = 10

graph = dict()
visited = dict()
distance = dict()
for i in range(1, n+1):
    graph[i] = list()
    visited[i] = False
    distance[i] = None

for v1, v2 in ipt:
    graph[v1].append(v2)
    graph[v2].append(v1)

SubTreeNodes(graph, 1, visited, distance)
print(distance)
```

```python
# Kruskal Algorithm

class Edge:
    def __init__(self, source, destination, weight):
        self.source = source
        self.destination = destination
        self.weight = weight

def getParent(v, parent):
    # Recursive function to find the parent of a vertex
    if v == parent[v]:
        return v
    return getParent(parent[v], parent)

def kruskal(edges, vertices):
    # Kruskal's algorithm to find minimum spanning tree
    parent = [i for i in range(vertices)]  # Initialize parent array

    edges = sorted(edges, key=lambda edge: edge.weight)  # Sort edges by weight

    count = 0  # Initialize count of edges added to MST
    output = []  # Initialize output list to store MST edges
    i = 0  # Initialize index for iterating over sorted edges
    while count < (vertices - 1):  # Loop until MST is formed
        currentEdge = edges[i]  # Get the current edge
        srcParent = getParent(currentEdge.source, parent)  # Find parent of source vertex
        destParent = getParent(currentEdge.destination, parent)  # Find parent of destination vertex

        if srcParent != destParent:  # If including this edge does not form a cycle
            output.append(currentEdge)  # Add the edge to the MST
            count += 1  # Increment count of edges added
            parent[srcParent] = destParent  # Update parent to merge the sets
        i += 1  # Move to the next edge
    return output  # Return the MST

# Input parsing
li = [int(ele) for ele in input().strip().split()]
vertices = li[0]  # Number of vertices
E = li[1]  # Number of edges
edges = []  # List to store edges

for i in range(E):
    curr_input = [int(ele) for ele in input().strip().split()]
    source = curr_input[0]  # Source vertex of the edge
    destination = curr_input[1]  # Destination vertex of the edge
    weight = curr_input[2]  # Weight of the edge
    edge = Edge(source, destination, weight)  # Create Edge object
    edges.append(edge)  # Add edge to the list

# Print edges array
print("Edges Array:")
for edge in edges:
    print("Source:", edge.source, "Destination:", edge.destination, "Weight:", edge.weight)

output = kruskal(edges, vertices)  # Find MST using Kruskal's algorithm
# Print MST edges
print("MST Edges:")
for edge in output:
    if edge.source < edge.destination:
        print(str(edge.source) + " --> " + str(edge.destination) + " == " + str(edge.weight))
    else:
        print(str(edge.destination) + " --> " + str(edge.source) + " == " + str(edge.weight))
```

```python
class Graph:
    def __init__(self, nVertices):
        """
        Initialize the graph with a given number of vertices.
        Create an adjacency matrix initialized with 0's.
        """
        self.nVertices = nVertices
        self.adjacentMatrix = [[0 for j in range(nVertices)] for i in range(nVertices)]

    def addEdges(self, v1, v2, wt):
        """
        Add an edge between two vertices with a given weight.
        Update the adjacency matrix to reflect the new edge.
        
        Parameters:
        v1 (int): The first vertex.
        v2 (int): The second vertex.
        wt (int): The weight of the edge.
        """
        self.adjacentMatrix[v1][v2] = wt
        self.adjacentMatrix[v2][v1] = wt

    def removeEdges(self, v1, v2):
        """
        Remove an edge between two vertices.
        Update the adjacency matrix to remove the edge.
        
        Parameters:
        v1 (int): The first vertex.
        v2 (int): The second vertex.
        """
        if not self.containsEdges(v1, v2):
            return
        self.adjacentMatrix[v1][v2] = 0
        self.adjacentMatrix[v2][v1] = 0

    def containsEdges(self, v1, v2):
        """
        Check if there is an edge between two vertices.
        
        Parameters:
        v1 (int): The first vertex.
        v2 (int): The second vertex.
        
        Returns:
        bool: True if there is an edge, False otherwise.
        """
        return self.adjacentMatrix[v1][v2] > 0

    def __str__(self):
        """
        Return the string representation of the adjacency matrix.
        """
        return str(self.adjacentMatrix)

    def printAdjMat(self):
        """
        Return the adjacency matrix.
        """
        return self.adjacentMatrix

    def __get_minVertex(self, visited, weight):
        """
        Find the vertex with the minimum weight that has not been visited.
        
        Parameters:
        visited (list of bool): List indicating if a vertex has been visited.
        weight (list of float): List of weights to vertices.
        
        Returns:
        int: The index of the vertex with the minimum weight.
        """
        min_vertex = -1
        for i in range(self.nVertices):
            if not visited[i] and (min_vertex == -1 or weight[min_vertex] > weight[i]):
                min_vertex = i
        return min_vertex

    def prims(self):
        """
        Implement Prim's Algorithm to find the Minimum Spanning Tree (MST).
        Print the edges and their weights in the MST.
        """
        visited = [False for i in range(self.nVertices)]  # Track visited vertices
        parentArray = [-1 for i in range(self.nVertices)]  # Track parent vertices
        weight = [float('inf') for i in range(self.nVertices)]  # Track edge weights
        weight[0] = 0  # Start with the first vertex

        # Iterate over the vertices to build the MST
        for _ in range(self.nVertices - 1):
            minVertex = self.__get_minVertex(visited, weight)
            visited[minVertex] = True

            # Update the weights and parents for the adjacent vertices
            for j in range(self.nVertices):
                if self.adjacentMatrix[minVertex][j] > 0 and not visited[j]:
                    if weight[j] > self.adjacentMatrix[minVertex][j]:
                        weight[j] = self.adjacentMatrix[minVertex][j]
                        parentArray[j] = minVertex

        # Print the edges in the MST along with their weights
        for i in range(1, self.nVertices):
            if parentArray[i] < i:
                print(f"{parentArray[i]} - {i} (Weight: {weight[i]})")
            else:
                print(f"{i} - {parentArray[i]} (Weight: {weight[i]})")

# Read the number of vertices and edges from input
V_E = [int(i) for i in input().strip().split()]
V = V_E[0]
E = V_E[1]

# Initialize the graph
g = Graph(V)

# Add edges to the graph
for _ in range(E):
    currentInput = [int(i) for i in input().strip().split()]
    g.addEdges(currentInput[0], currentInput[1], currentInput[2])

# Execute Prim's Algorithm and print the MST
g.prims()
```

```python
from heapq import heappop, heappush

def prims(graph, start, parent, distance, visited):
    bag = []
    heappush(bag, (0, start))  # Use tuples (distance, node)
    distance[start] = 0
    parent[start] = -1

    while bag:
        dist, node = heappop(bag)
        if not visited[node]:
            visited[node] = True
            for childNode, childDist in graph[node]:
                if not visited[childNode] and distance[childNode] > childDist:
                    parent[childNode] = node
                    distance[childNode] = childDist
                    heappush(bag, (childDist, childNode))

    print("Distance array:", distance)
    print("Parent array:", parent)

    # Print the edges and their weights in the MST
    print("Minimum Spanning Tree Edges and Weights:")
    for node in range(1, len(parent)):
        if parent[node] != -1:  # Skip the starting node with no parent
            print(f"{parent[node]} - {node} (Weight: {distance[node]})")

# Test case
# ipt = [[0, 1, 3], [0, 3, 5], [1, 2, 1], [2, 3, 8]]
# n = 4

ipt = [
    [0, 1, 4], 
    [0, 2, 2], 
    [1, 2, 5], 
    [1, 3, 10], 
    [2, 3, 3]
]
n = 4

graph = {i: [] for i in range(n)}
parent = {i: None for i in range(n)}
distance = {i: float('inf') for i in range(n)}
visited = {i: False for i in range(n)}

for v1, v2, wt in ipt:
    graph[v1].append((v2, wt))  # Use tuples
    graph[v2].append((v1, wt))

start = 0
prims(graph, start, parent, distance, visited)
```
